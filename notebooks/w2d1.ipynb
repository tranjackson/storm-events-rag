```python
from sentence_transformers import SentenceTransformer

from storm_rag.geo import load_all_events

import numpy as np

model = SentenceTransformer("BAAI/bge-small-en-v1.5")  # first run downloads ~130MB

ev = load_all_events().dropna(subset=["EVENT_NARRATIVE"]).head(500).reset_index(drop=True)

emb = model.encode(ev["EVENT_NARRATIVE"].tolist(), normalize_embeddings=True)
```

```python
q = model.encode(["severe flash flooding damaged homes overnight"], normalize_embeddings=True)

scores = (emb @ q.T).ravel()

top = np.argsort(-scores)[:5]

for i in top:
    print(round(float(scores[i]), 3), ev.loc[i, "EVENT_NARRATIVE"][:120])
```
